### Data Cleaning:

```python
#  Missing Values & Duplicates
print("\n--- Data Cleaning ---")
if df.isnull().sum().sum() == 0:
    print("No missing values found.")
if df.duplicated().sum() == 0:
    print("No duplicate entries found.")
```

```python
#  Outlier Identification (Price)
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['Price'])
plt.title('Outlier Check: House Price Distribution')
plt.xlabel('Price ($)')
plt.show()
```


### Outlier Treatment 

1.  **Identification:** The boxplot showed many data points that were extremely high (outliers) in the house price range.
2.  **Decision:** We chose **NOT to remove** these outliers.
    * **Reason:** In a house price dataset, these represent real, high-value luxury homes. They are valid market data, not errors.
3.  **Mitigation:** Since the outliers distort the standard average (the Mean), we used **robust statistics** instead:
    * We reported the **Median Price**, which is not affected by outliers.
    * We calculated the **Trimmed Mean**, which throws out the highest and lowest 10% of prices before calculating the average, giving a more stable result.

### UNIVARIATE ANALYSIS 

```python
# Calculate Measures of Central Tendency and Variability

from scipy.stats import trim_mean

numerical_cols = ['Price', 'living area']
stats = {}

for col in numerical_cols:
    data = df[col]
    stats[col] = {
        'Mean': data.mean(),
        'Median': data.median(),
        'Trimmed Mean (10%)': trim_mean(data, proportiontocut=0.1), 
        'Standard Deviation': data.std(),
        'Range': data.max() - data.min()
    }
stats_df = pd.DataFrame(stats).applymap('{:,.2f}'.format)
print("\nUnivariate Statistics:")
print(stats_df)
```

```python
#  Visualize Numerical Distributions (Histogram)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['Price'], kde=True, bins=50)
plt.title('2a. Price Distribution')

plt.subplot(1, 2, 2)
sns.histplot(df['living area'], kde=True, bins=50, color='coral')
plt.title('2b. Living Area Distribution')
plt.tight_layout()

plt.show()
```

```python
# Categorical Frequencies (Condition)
condition_counts = df['condition of the house'].value_counts().sort_index()
print("\nCondition of the House Frequencies:\n", condition_counts)


plt.figure(figsize=(6, 4))
sns.countplot(x='condition of the house', data=df, order=condition_counts.index, palette='viridis')
plt.title('3. Frequency of House Condition Grade')
plt.xlabel('Condition Grade (1: Poor to 5: Excellent)')

plt.show()
```

### BIVARIATE/MULTIVARIATE ANALYSIS

```python
#  Correlation Heatmap
corr_cols = ['Price', 'living area', 'number of bathrooms', 'grade of the house', 'condition of the house']
corr_matrix = df[corr_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('4. Correlation Matrix Heatmap')

plt.show()
```

```python
#  Scatter Plot (Price vs. Living Area)
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['living area'], y=df['Price'], alpha=0.6)
plt.title('5. Price vs. Living Area Relationship')
plt.xlabel('Living Area (sq ft)')
plt.ylabel('Price ($)')

plt.show()
```

```python
#  Grouped Boxplot (Price vs. Condition)
plt.figure(figsize=(8, 6))
sns.boxplot(x='condition of the house', y='Price', data=df, palette='Spectral')
plt.title('6. Price Distribution by House Condition Grade')
plt.xlabel('Condition Grade')
plt.ylabel('Price ($)')
plt.show()
```

### Insights:
1. The housing market is highly skewed by expensive, luxury outliers.

2. House Quality (Grade) is the strongest factor driving price, even more than size.

3. 	House size (living area) strongly and positively affects price, as expected.

4. Better house condition increases the typical (median) price.

5. 	The number of bathrooms is a significant value driver for buyers.
